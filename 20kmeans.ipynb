```python
import numpy as np

k = 4
n = 200
data = np.random.randn(n, 2)

centroids = data[np.random.choice(np.arange(n), size=(k,))]
print(centroids)

for l in range(10):
  indexes = np.zeros(data.shape[0])
  for centroid in centroids:  # <--この行いらない
    for i, x in enumerate(data):
      if i == 0:
        print(np.sum((x - centroids) ** 2, axis=1))
      indexes[i] = np.argmin(np.sum((x - centroids) ** 2, axis=1))
  
  for i in range(k):
    centroids[i] = data[indexes == i].mean(axis=0)

print(centroids)
```
