```python
import numpy as np

n = 100
k = 4
data = np.random.randn(n, 2)
prob = np.repeat(1/n, n)
print(prob)
centroids = np.zeros((k, 2))
distances = np.zeros((n, k))

for i in range(k):
  centroids[i] = data[np.random.choice(np.arange(n), p=prob, size=(1))]
  distances[:, i] = np.sum((data - centroids[i])**2, axis=1)
  prob = np.sum(distances, axis=1) / np.sum(distances)
  print(distances)
  print(prob)
  print(len(prob))
```
